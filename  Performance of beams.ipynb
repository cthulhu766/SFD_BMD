# Reaction types (Типы реакций)
Балки обычно опираются на опоры или колонны, которые обеспечивают необходимое равновесие. (Согласно третьему закону Ньютона, на каждое действие существует равная и противоположная реакция). Проще говоря, нагрузка на балку - это действие, а опоры обеспечивают реакцию. Существует три вида реакции опор:


# Loading types (Типы нагрузки)
Балки подвергаются равномерно распределенным нагрузкам (UDL), точечным (сосредоточенным) нагрузкам или их комбинации. Ниже показаны различные условия нагружения, которым может подвергаться балка.

\
**Sum of the vertical forces must be zero	Σ Fv = 0**\
**Sum of the horizontal forces must be zero	Σ Fh = 0** \
**Sum of the moments forces must be zero	Σ M = 0** \
\
Перерезывающая сила в любом сечении балки равна алгебраической сумме внешних сил, действующих только на одну сторону сечения.

Изгибающий момент в любой точке балки равен алгебраической сумме моментов (взятых относительно точки) внешних сил (нагрузок и реакций), действующих только на одну сторону сечения.

# Example 1 - Beam with a uniform distributed load


```python
import numpy as np
import matplotlib.pyplot as plt

l = 8 # length of beam
w = 3000 # N/m
```

```python
# Reactions
R = w*l/2
x = np.linspace(0,l,80) 

X = []
SF = []
M = []

# Добавляем данные в список
for y in x:
    Mx = (R*y) - w*(y**2/2)
    sf = R - (w*y)
    X.append(y)
    M.append(Mx)
    SF.append(sf)
plt.figure(figsize=(7.5,5),dpi=100)   #dpi - dots per inch
plt.subplot(2,1,1) # 2 графика, одна колонна, 1-й график
plt.plot(X,SF,'g')
plt.fill_between(X,SF,color="green",alpha=0.5,hatch="||") #alpha - создает дополнительные диления на "зелёном" фоне
plt.title("Shear Force Diagram")
plt.xlabel("Length")
plt.ylabel("Shear Force [N]")
plt.tight_layout(pad=4.0)

plt.subplot(2,1,2) # 2 графика, одна колонна, 2-й график
plt.plot(X,M,'r')
plt.fill(X,M, color='red',alpha=0.5,hatch="||")
plt.title("Bending Moment Diagram")
plt.xlabel("Length")
plt.ylabel("Bending Moment [Nm]")
plt.show()
```
